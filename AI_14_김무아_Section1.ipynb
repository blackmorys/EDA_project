```python
df['NA_Sales'] = df['NA_Sales'].str.replace(pat=r'[M]', repl=r'', regex = True)
df['EU_Sales'] = df['EU_Sales'].str.replace(pat=r'[M]', repl=r'', regex = True)
df['JP_Sales'] = df['JP_Sales'].str.replace(pat=r'[M]', repl=r'', regex = True)
df['Other_Sales'] = df['Other_Sales'].str.replace(pat=r'[M]', repl=r'', regex = True)
```

```python
def fix_columns(col):
    temp_list = []

    for i in col:
        temp = float(i.replace("K",""))
        temp_list.append(temp)

    return temp_list

les_na = df.loc[df.NA_Sales.str.contains('K')]
les_na
```

```python
les_na['NA_Sales'] = fix_columns(les_na['NA_Sales'])
les_na['NA_Sales'] = les_na['NA_Sales'] * 0.001
les_na = les_na['NA_Sales']
les_na
```

```python
les_eu = df.loc[df.EU_Sales.str.contains('K')]
les_eu
```

```python
les_eu['EU_Sales'] = fix_columns(les_eu['EU_Sales'])
les_eu['EU_Sales'] = les_eu['EU_Sales'] * 0.001
les_eu = les_eu['EU_Sales']
les_eu
```

```python
les_jp = df.loc[df.JP_Sales.str.contains('K')]
les_jp
```

```python
les_jp['JP_Sales'] = fix_columns(les_jp['JP_Sales'])
les_jp['JP_Sales'] = les_jp['JP_Sales'] * 0.001
les_jp = les_jp['JP_Sales']
les_jp
```

```python
les_ot = df.loc[df.Other_Sales.str.contains('K')]
les_ot
```

```python
les_ot['Other_Sales'] = fix_columns(les_ot['Other_Sales'])
les_ot['Other_Sales'] = les_ot['Other_Sales'] * 0.001
les_ot = les_ot['Other_Sales']
les_ot
```

```python
df['NA_Sales'].loc[df.NA_Sales.str.contains('K')] = les_na
df['EU_Sales'].loc[df.EU_Sales.str.contains('K')] = les_eu
df['JP_Sales'].loc[df.JP_Sales.str.contains('K')] = les_jp
df['Other_Sales'].loc[df.Other_Sales.str.contains('K')] = les_ot
df
```

```python
df = df.astype({'NA_Sales':'float', 'EU_Sales':'float', 'JP_Sales':'float', 'Other_Sales':'float'})
df.sum()
```

**2.연도별 게임의 트렌드가 있을까**

```python
df.groupby(['Year', 'Genre']).size()
```

```python
y_pro = df.loc[df.Year < 1700]
y_19 = y_pro.loc[y_pro.Year > 50].index
y_20 = y_pro.loc[y_pro.Year < 50].index
y_pro.loc[y_19, 'Year'] += 1900
y_pro.loc[y_20, 'Year'] += 2000
y_pro
```

```python
y_index = df.loc[df.Year < 1700].index
df.loc[y_index, 'Year'] = y_pro.Year
df.loc[y_index]
```

```python
df.groupby(['Year', 'Genre']).size()
```

```python
q2 = df.groupby(['Year', 'Genre']).size().reset_index(name = 'count')
q2_count = q2.groupby(['Year'])['count'].transform(max) == q2['count']
q2_genre = q2[q2_count].reset_index(drop = True)
q2_count
q2_genre = q2_genre.drop_duplicates(subset = ['Year','count'], keep = 'last').reset_index(drop = True)
q2_genre
```

```python
plt.figure(figsize = (30,20))
q2_gr = sns.barplot(x = 'Year', y = 'count', data = q2_genre)
q2_gr.set_title('Top-selling genre by year', fontsize = 30)
i = 0
for value in q2_genre['count']:
    q2_gr.text(x = i, y = value + 5, s = str(q2_genre.Genre[i]), rotation = 90)
    i += 1
plt.show()
```

1990년대에 스포츠 장르가 유행을 했지만 대부분 액션 장르가 트렌드였다.

**3.출고량이 높은 게임에 대한 분석 및 시각화 프로세스**

```python
df['Sales_s'] = df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales']
q3 = df.sort_values('Sales_s', ascending = False)
q3 = q3.head(10)
q3
```

```python
plt.figure(figsize = (30,15))
q3_gr = sns.barplot(x = 'Name', y='Sales_s', data = q3)
q3_gr.set_title('Top 10 Selling Games', fontsize = 30)
plt.show()
```

판매량 TOP 10의 게임들은 모두 닌텐도에서 발매 한 게임들이였다.
그 중에선 강력한 IP(마리오,포켓몬)의 힘을 가진 게임,
파티게임(로컬플레이,쉬운 규칙, 같은 조건) 장르라고 불리기도 하는 게임,
고전명작 게임으로 구분이 되었다.
